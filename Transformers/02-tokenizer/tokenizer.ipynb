```python
from transformers import AutoTokenizer
```

```python
sen = "弱小的我也有大梦想！"
```

```python
# 1. AutoTokenizer的导入和保存
tokenizer = AutoTokenizer.from_pretrained("uer/roberta-base-finetuned-dianping-chinese")
tokenizer
# tokenizer 保存到本地
tokenizer.save_pretrained("./uer/roberta-base-finetuned-dianping-chinese")
```

```python
# tokenizer的本地加载
tokenizer = AutoTokenizer.from_pretrained("./uer/roberta-base-finetuned-dianping-chinese")
tokenizer

```

```python
# 2 句子分词
tokens = tokenizer.tokenize(sen)
tokens
```

```python
# 查看词典
tokenizer.vocab
```

```python
tokenizer.vocab_size
```

```python
# 4 索引转换
ids = tokenizer.convert_tokens_to_ids(tokens)
ids
```

```python
# 4 索引转换
tokens = tokenizer.convert_ids_to_tokens(ids)
tokens
```

```python
# 5 更便捷的处理方式：encode
ids = tokenizer.encode(sen)
ids


```

```python
## 将ids转换为字符串，又称为解码
str_sen =tokenizer.decode(ids)
print(str_sen)

str_sen = tokenizer.decode(ids, skip_special_tokens=True)
print(str_sen)

##  截断
ids = tokenizer.encode(sen, max_length=10, truncation=True)
print(ids)

# 6 填充
ids = tokenizer.encode(sen, max_length=15, padding="max_length")
print(ids)
```

```python
# 6 其他
attention_mask = [1 if idx !=0 else 0 for idx in ids]
token_type_ids = [0]*len(ids)
ids, attention_mask, token_type_ids


```

```python
# 7 快速调用方式
## 使用 tokenizer.encode_plus 快速调用
inputs = tokenizer.encode_plus(sen,padding="max_length",max_length=15)
inputs
```

```python
## 直接使用tokenizer 的快速调用方式
inputs = tokenizer(sen,padding="max_length",max_length=15)
inputs
```

```python
# 8 处理 batch 数据
sens =[ "I love you", "I love you too","追逐梦想的心，比梦想本身，更可贵！"]
res = tokenizer(sens)
res
```

```python
# 9 Fast/Slow tokenizer
sen = "弱小的我也有大 Dreaming"
fast_tokenizer = AutoTokenizer.from_pretrained("uer/roberta-base-finetuned-dianping-chinese")
fast_tokenizer
```
